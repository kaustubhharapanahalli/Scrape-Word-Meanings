```python
import time
import csv
from pprint import pprint

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
```

```python
chromeOptions = Options()
chromeOptions.add_argument("--ignore-certificate-errors")
chromeOptions.add_argument("--incognito")
chromeOptions.add_argument("--headless")
# chromeOptions.add_argument("--kiosk") # Maximize window size
```

```python
list_of_words_grouped = list()
with open("GRE_Words.csv", "r", newline="") as file:
    reader = csv.reader(file)
    for row in reader:
        list_of_words_grouped.append(row)
```

```python
for word_list in list_of_words_grouped:
    while word_list[-1] == "":
        del word_list[-1]
```

```python
print(list_of_words_grouped)
```

```python
consolidated_json = {}

for group_number in range(len(list_of_words_grouped)):
    name = list_of_words_grouped[group_number][0]
    print(name)
    consolidated_json[name] = {}
    for word in list_of_words_grouped[group_number][1:]:
        consolidated_json[name][word] = {}
        driver = webdriver.Chrome(r"D:\Downloads\chromedriver_win32\chromedriver.exe", options=chromeOptions)
        driver.get("https://www.lexico.com/")
        
        time.sleep(1)
        change_dict = driver.find_element_by_class_name("sbSelector")
        change_dict.click()

        time.sleep(1)
        
        select_dict = driver.find_element_by_class_name("sbOptions").find_element_by_link_text("US DICTIONARY")
        select_dict.click()
        try:
            elem = driver.find_element_by_id("query")
        except Exception:
            time.sleep(900)
            elem = driver.find_element_by_id("query")
        time.sleep(1)
        elem.clear()
        elem.send_keys(word)
        elem.send_keys(Keys.RETURN)
        print(f"Navigated to the word: {word}")

        time.sleep(5)

        definitions = []
        try:
            def_elems = driver.find_element_by_class_name('trg').find_elements_by_class_name('ind')
        except Exception:
            driver.quit()
            time.sleep(2)
            
            driver = webdriver.Chrome(r"D:\Downloads\chromedriver_win32\chromedriver.exe", options=chromeOptions)
            driver.get("https://www.lexico.com/")

            time.sleep(1)
            change_dict = driver.find_element_by_class_name("sbSelector")
            change_dict.click()

            time.sleep(1)

            select_dict = driver.find_element_by_class_name("sbOptions").find_element_by_link_text("US DICTIONARY")
            select_dict.click()
            elem = driver.find_element_by_id("query")
            time.sleep(1)
            elem.clear()
            elem.send_keys(word)
            elem.send_keys(Keys.RETURN)
            print(f"Navigated to the word: {word}")
            time.sleep(5)
            def_elems = driver.find_element_by_class_name('trg').find_elements_by_class_name('ind')
            
        for elem in def_elems:
            definitions.append(elem.text)
        consolidated_json[name][word]["definition"] = definitions
        print(f"\tFetched the definitions of the word")

        try:
            def_elems = driver.find_element_by_class_name('trg').find_elements_by_class_name('moreInfo')
            for elem in def_elems:
                elem.click()
                time.sleep(2)
                
            example_sentences = []
            def_elems = driver.find_element_by_class_name('trg').find_elements_by_class_name('english-ex')
            for elem in def_elems:
                examples = elem.text
                example_sen = examples.split("\n")
                example_sentences.extend(example_sen)
            consolidated_json[name][word]["example_sentences"] = example_sentences
            print(f"\tFetched examples for the word")
            time.sleep(1)
            driver.quit()

        except Exception:
            try:
                driver.quit()
                time.sleep(2)

                driver = webdriver.Chrome(r"D:\Downloads\chromedriver_win32\chromedriver.exe", options=chromeOptions)
                driver.get("https://www.lexico.com/")

                time.sleep(1)
                change_dict = driver.find_element_by_class_name("sbSelector")
                change_dict.click()

                time.sleep(1)

                select_dict = driver.find_element_by_class_name("sbOptions").find_element_by_link_text("US DICTIONARY")
                select_dict.click()
                elem = driver.find_element_by_id("query")
                time.sleep(1)
                elem.clear()
                elem.send_keys(word)
                elem.send_keys(Keys.RETURN)
                print(f"Navigated to the word: {word}")
                time.sleep(5)
                def_elems = driver.find_element_by_class_name('trg').find_elements_by_class_name('ind')

                for elem in def_elems:
                    definitions.append(elem.text)
                consolidated_json[name][word]["definition"] = definitions
                print(f"\tFetched the definitions of the word")
                def_elems = driver.find_element_by_class_name('trg').find_elements_by_class_name('moreInfo')

                for elem in def_elems:
                    elem.click()
                    time.sleep(2)
                    
                example_sentences = []
                def_elems = driver.find_element_by_class_name('trg').find_elements_by_class_name('english-ex')
                for elem in def_elems:
                    examples = elem.text
                    example_sen = examples.split("\n")
                    example_sentences.extend(example_sen)
                consolidated_json[name][word]["example_sentences"] = example_sentences
                print(f"\tFetched examples for the word")
                time.sleep(1)
                driver.quit()

            except Exception:
                print("Unable to fetch example sentences")

        time.sleep(1)
        
        driver = webdriver.Chrome(r"D:\Downloads\chromedriver_win32\chromedriver.exe", options=chromeOptions)
        driver.get("https://www.lexico.com/synonyms")
        time.sleep(2)
        elem = driver.find_element_by_id("query")
        time.sleep(1)
        elem.clear()
        elem.send_keys(word)
        elem.send_keys(Keys.RETURN)
        print(f"Navigated to the word: {word}")

        time.sleep(2)
        synonyms_list = []
        synonyms = driver.find_elements_by_class_name("synGroup")
        for syn in synonyms:
            synonyms_list.extend(syn.text.split(", "))
        consolidated_json[name][word]["synonyms"] = synonyms_list
        print(f"\tFetched synonyms for the word")

        time.sleep(2)
        antonyms_list = []
        antonyms = driver.find_elements_by_class_name("antGroup")
        for syn in antonyms:
            if "," in syn.text:
                antonyms_list.extend(syn.text.split(", "))
            else:
                antonyms_list.append(syn.text)
        consolidated_json[name][word]["antonyms"] = antonyms_list
        print(f"\tFetched antonyms for the word")

        time.sleep(2)
        driver.quit()
        print("-" * 100)
    print("="*100)
    print(f"{name} completed")
    print("="*100)
    time.sleep(60)
```
